```python
import sys, os
sys.path.append(os.path.abspath('..'))
sys.path.append("../..")
from imports import ilpgraph, readFile
import xlsxwriter
from imports import networkx as imp_nx
#from network import Steiner_Flow as sflow
from network import Steiner_Linear_with_Flow as sflow
#from network import Steiner_Flow_withLinear as sflow
import networkx as nx
from gurobipy import *
import random

path =  'MVC_testInstances/cc6-3p.stp'
G, terminals = readFile.stp_to_networkx(path)
nodeList = list(G.nodes())
root = len(G.nodes()) + 1
accessPoints = [root + 1, root + 2, root + 3, root + 4]

newEdges = []

for ap in accessPoints:
    newEdges.append((root, ap))
    randomPoints = random.choices(nodeList, k=20)
    for rp in randomPoints:
        newEdges.append((ap, rp))
        
for node in G.nodes():
    if node in terminals:
        G.nodes[node]['Weight'] = random.randrange(1, 3)
    else:
        G.nodes[node]['Weight'] = 0

for edge in G.edges():
    G[edge[0]][edge[1]]['weight'] = random.randrange(50,100)
    G[edge[0]][edge[1]]['Capacity'] = random.randrange(300,600)


G.add_node(root, Weight = 0)
G.add_nodes_from(accessPoints, Weight = 0)
G.add_edges_from(newEdges, weight = 0, Capacity = 100000000)
print("Connected: ", nx.is_connected(G))
print("Nodes: ", len(G.nodes()))
print("Edges: ", len(G.edges()))
print("Root: ", root)
print("Terminals: ", terminals)

Graph = imp_nx.read(G)
model = sflow.createModel(Graph, terminals, root, 'weight')
model.params.MIPGap = 0.05
model.params.TimeLimit = 1500
model.optimize()


```
