```python
import sys, os
sys.path.append(os.path.abspath('..'))
sys.path.append("../..")
from imports import ilpgraph, readFile
import xlsxwriter
from imports import networkx as imp_nx
#from network import Steiner_Flow as sflow
from network import Steiner_Linear_with_Flow as sflow
#from network import Steiner_Flow_withLinear as sflow
import networkx as nx
from gurobipy import *
import random

def __write_solution(solution:list):
    
    workbook = xlsxwriter.Workbook('BenchmarksLinear_Flow.xlsx')
    worksheet = workbook.add_worksheet()
    bold = workbook.add_format({'bold':True})
    solutionRows = len(solution)
    solutionColumns = len(solution[0])
    
    worksheet.write('A1', 'Instance', bold)
    worksheet.write('B1', 'Edges', bold)
    worksheet.write('C1', 'Nodes', bold)
    worksheet.write('D1', 'Terminals', bold)
    worksheet.write('E1', 'Runtime', bold)
    worksheet.write('F1', 'Objective Value', bold)
    worksheet.write('G1', 'Constraints', bold)
    worksheet.write('H1', 'MIP Gap', bold)
    
    print(solution)
    for i in range(solutionRows):
        for j in range(solutionColumns):
            worksheet.write(i + 1, j, solution[i][j])
    
        
def runBenchmark(paths:list):
    solutions = []
    i = 0
    
    for path in paths:
        with open(path) as f:
            G, terminals = readFile.stp_to_networkx(path)
            for edge in G.edges():
                G[edge[0]][edge[1]]['Cost'] = 1

            root = terminals[3]
            print("Connected: ", nx.is_connected(G))
            print("Nodes: ", len(G.nodes()))
            print("Edges: ", len(G.edges()))
            print("Root: ", root)
            print("Terminals: ", terminals)
            Graph = imp_nx.read(G)
            model = sflow.createModel(Graph, terminals, 'weight', root)
            model.params.MIPGap = 0.05
            model.params.TimeLimit = 1500
            model.optimize()
            solutions.append((i, len(G.edges()), len(G.nodes()), len(terminals), model.runTime, model.objVal,\
              model.numConstrs, model.MIPGap))
            i += 1

    __write_solution(solutions)         

if __name__ == '__main__':
    

    paths = ['MVC_testInstances/d09.stp',
             'MVC_testInstances/d13.stp',
             'MVC_testInstances/e08.stp',
             'MVC_testInstances/e10.stp',
             'MVC_testInstances/e13.stp',
             'MVC_testInstances/e15.stp',
             'MVC_testInstances/e20.stp',
             'MVC_testInstances/cc6-3p.stp',
             'MVC_testInstances/bipe2p.stp']
    
    runBenchmark(paths)
    


```
